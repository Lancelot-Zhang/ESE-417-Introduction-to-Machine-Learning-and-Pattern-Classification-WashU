```python
# this Python code is an example of using the linear regression model from sklearn.
# no train and test data set split is used.

import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error


# Let's assume the truth behind the given data set is a sin function (1D example)
# y = Asin(x) + ax + b + noise

Int_t = 10           #intercept of the linear function
Slope_t = 0.1        #slope of the linear function

mu, sigma = 0, 0.1   # Here are the parameters of the normal noise

N_sample = 50        #number of sampled data points
mag_sin = 0.6        #the magnitude of the sin function

rnd = np.random.normal(mu,sigma,N_sample)

x = np.linspace(0,10,N_sample)                #the inputs
y_t = mag_sin*np.sin(x) + x*Slope_t + Int_t   #the true outputs
y = y_t + rnd                                 #the target values

x_m = x[:, np.newaxis]  # convert x into a matrix
lrg = linear_model.LinearRegression()  # create a linear regression object 
lrg.fit(x_m,y)                         # fit the linear model using the synthetic data set
y_hat = lrg.predict(x_m)               # make prediction on the inputs using the fitted linear regression model

SSE = mean_squared_error(y,y_hat)*N_sample  # the sum of squared error of the fitted linear squared line

print("the intercep is: {}\n".format(lrg.intercept_))
print("the slope is: {}\n".format(lrg.coef_))
print("SSE = {}".format(SSE))


##display the least-squared fitted line

plt.scatter(x,y,c="k",s=5,label="data points")
plt.plot(x, y_hat, 'r',label="least square line")
plt.plot(x, y_t, "g",label="the truth")
plt.title("the truth, data points and the least squared line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()


```
